```python
import pandas as pd
import numpy as np


df = pd.read_csv('../data/raw/superstore.csv', encoding='latin-1')
print(df.shape)          # Debería dar (9994, 21)
print(df.dtypes)
df.head(3)


print(df.isnull().sum())

# Verificar duplicados
print(f"Duplicados: {df.duplicated().sum()}")

# Estadísticas descriptivas de columnas numéricas clave
df[["Sales", "Quantity", "Discount", "Profit"]].describe()



df["Postal Code"] = df["Postal Code"].fillna("00000")
df["Postal Code"] = df["Postal Code"].astype(str).str.zfill(5)



df_clean = df.drop_duplicates()
print(f"Filas originales: {len(df)} | Filas después: {len(df_clean)}")


df_clean["Order Date"] = pd.to_datetime(df_clean["Order Date"], format="%m/%d/%Y")
df_clean["Ship Date"]  = pd.to_datetime(df_clean["Ship Date"],  format="%m/%d/%Y")


invalid_dates = df_clean[df_clean["Ship Date"] < df_clean["Order Date"]]
print(f"Fechas incoherentes: {len(invalid_dates)}")



df_clean.columns = [
    "row_id", "order_id", "order_date", "ship_date", "ship_mode",
    "customer_id", "customer_name", "segment", "country", "city",
    "state", "postal_code", "region", "product_id", "category",
    "sub_category", "product_name", "sales", "quantity", "discount", "profit"
]



# Margen de ganancia por fila
df_clean["profit_margin"] = (df_clean["profit"] / df_clean["sales"]).round(4)

# Año y mes de la orden (facilitará el análisis temporal en SQL/Power BI)
df_clean["order_year"]  = df_clean["order_date"].dt.year
df_clean["order_month"] = df_clean["order_date"].dt.month

# Días de envío (lead time)
df_clean["shipping_days"] = (df_clean["ship_date"] - df_clean["order_date"]).dt.days

# Flag booleano de orden con pérdida
df_clean["is_loss"] = df_clean["profit"] < 0



# Revisión final de la estructura
print(df_clean.dtypes)
print(df_clean.isnull().sum())
print(df_clean.shape)

# Exportar CSV limpio listo para la base de datos
df_clean.to_csv('../data/processed/superstore_clean.csv', index=False)
print("Dataset limpio exportado correctamente ✓")
```
